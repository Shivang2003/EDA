Q6.Compare different delivery rating counts of different restaurants.
Also tell that which restaurants has got the highest rating counts.
Do this by creating a barplot

```python
restaurant_rating_counts = zr['Restaurant_Name'].value_counts()

for i in range(10):
  plt.figure(figsize=(12, 6))
  restaurant_rating_counts[i:i+10].plot(kind='bar', color='darkblue')
  plt.xlabel('Restaurant Name')
  plt.ylabel('Delivery Rating Counts')
  plt.title('Delivery Rating Counts for Different Restaurants ')
  plt.show()
```
